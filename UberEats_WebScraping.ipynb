```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

url = "https://www.ubereats.com/search?diningMode=DELIVERY&pl=JTdCJTIyYWRkcmVzcyUyMiUzQSUyMkJvc3RvbiUyMENvbW1vbiUyMiUyQyUyMnJlZmVyZW5jZSUyMiUzQSUyMjljZjNjZjU2LTc2ZTQtNDFmNC02MDU4LTBhODNjOGU3OTI0OSUyMiUyQyUyMnJlZmVyZW5jZVR5cGUlMjIlM0ElMjJ1YmVyX3BsYWNlcyUyMiUyQyUyMmxhdGl0dWRlJTIyJTNBNDIuMzU1MTQ0MSUyQyUyMmxvbmdpdHVkZSUyMiUzQS03MS4wNjU2OTQ5JTdE&q=Pizza&sc=SHORTCUTS"

TIMEOUT = 5

options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-notifications")
options.add_argument("--disable-popup-blocking")
options.add_argument("--disable-infobars")
options.add_argument("--disable-extensions")
options.add_experimental_option("prefs", 
                                {"profile.default_content_setting_values.notifications": 2 
                                }) 

print(f"Retrieving web page URL '{url}'")
driver = webdriver.Chrome(options=options)
driver.get(url)

# Timeout needed for the web page to render (read more about it)
time.sleep(TIMEOUT)

# Add a delay to allow the page to finish loading (adjust as needed)
time.sleep(TIMEOUT)

html = driver.page_source

# Get the source of the page and create a BeautifulSoup object
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())

# Close the browser
driver.quit()

print(f"Done")
```

```python
# Find all 'div' elements
restaurant_elements = soup.find_all('div', {'class': 'bo gn d1 br bh bw bu dd'})

# Extract restaurant names and store them in a list
restaurants = [res.text.strip() for res in restaurant_elements]
print(restaurants)
print(len(restaurants))
```

```python
div_tags_with_span = soup.find_all('div', {'class': 'al am fz iv iy'})

rating = []
total_reviews = []

for div in div_tags_with_span:
    span = div.find('span', {'class': 'ag bs bt bu bv bw bx'})
    if span is not None:
        review_text = span.text.strip()
        # Use regular expressions to extract the rating and total reviews
        import re
        match = re.search(r'(\d+\.\d+)\s+out of 5 stars based on (\d+) reviews', review_text)
        if match:
            rating.append(float(match.group(1)))  # Convert the rating to a float
            total_reviews.append(int(match.group(2)))  # Convert the total reviews to an int
        else:
            rating.append("not provided")
            total_reviews.append("not provided")
    else:
        rating.append("not provided")
        total_reviews.append("not provided")

print("Ratings:", rating)
print("Total Reviews:", total_reviews)
print(len(rating))
print(len(total_reviews))
```

```python
# Find all 'span' elements with reviews
time_elements = soup.find_all('div', {'class': 'bo d0 bq di mo'})

# Extract restaurant names and store them in a list
delivery_time = [time_span.text.strip() for time_span in time_elements]
print(delivery_time)
print(len(delivery_time))
```

```python
df = pd.DataFrame({
    'Restaurant Name': restaurants,
    'Restaurant Rating': rating,
    'Totsl Number of Reviews': total_reviews,
    'Delivery Time': delivery_time
})

csv_file = 'restaurant_data.csv'

df.to_csv(csv_file, index=False)

print(f'CSV file "{csv_file}" created successfully.')
```
